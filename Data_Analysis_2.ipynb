```python
import pandas as pd

# Display descriptive statistics
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set_style("whitegrid")

# Create the histogram
plt.figure(figsize=(10, 5))
sns.histplot(df["Social_Media_Followers"], bins=20, kde=True, color="blue")

# Labels and title
plt.xlabel("Social Media Followers")
plt.ylabel("Frequency")
plt.title("Distribution of Social Media Followers")

# Show the plot
plt.show()
```
